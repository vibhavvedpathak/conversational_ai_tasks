```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
```

```python
with open('JaneEyre.txt', 'r') as file:
  text = file.read()
```

```python
wordcloud = WordCloud(
width=800,
height=400,
background_color="white",
stopwords=set(STOPWORDS),
min_font_size=10,
).generate(text)
```

```python
plt.figure(figsize=(8, 8), facecolor=None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt


# Customize the WordCloud
wordcloud = WordCloud(
    width=800,
    height=400,
    background_color="black",  # Set your desired background color
    colormap="plasma",          # Experiment with different colormaps like 'viridis', 'plasma', 'inferno', etc.
    max_words=100,              # Change the maximum number of words
    stopwords=STOPWORDS
).generate(text)

# Display the WordCloud
plt.figure(figsize=(8, 8), facecolor=None)
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
from PIL import Image
import numpy as np

# Assuming 'upvote.png' is in the current working directory
mask_image = np.array(Image.open('upvote.png'))
```

```python
wordcloud = WordCloud(
    width=800,
    height=400,
    background_color="white",
    mask=mask_image,
    contour_color="steelblue",
    contour_width=1,
    stopwords=STOPWORDS
).generate(text)
```

```python
plt.figure(figsize=(8, 8), facecolor=None)
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
